```python
# ============================================
# SIMULADOR DE INTELIGENCIA ARTIFICIAL
# Tema: Investigador Científico
# Uso de if, elif, else y operadores
# Sin funciones, sin ciclos y sin POO
# ============================================

lista_vacia = []

print("====================================")
print(" IA INVESTIGADORA CIENTÍFICA ")
print("====================================")

while True:
    print("\nHola investigador.")
    nombre = input("Ingresa tu nombre: ")

    print("\nSelecciona el área de investigación")
    print("1. Biología")
    print("2. Tecnología")
    print("3. Espacio")

    area = input("Opción: ")

    # -------------------------
    # ÁREA BIOLOGÍA
    # -------------------------
    if area == "1":
        print("\nÁrea seleccionada: Biología")

        muestras = int(input("¿Cuántas muestras encontraste?: "))
        temperatura = int(input("Temperatura del laboratorio: "))

        # Uso de operadores lógicos
        if muestras >= 10 and temperatura <= 25:
            print("\nIA:", nombre + ", el experimento fue EXITOSO.")
            print("Las muestras son estables para análisis.")
            lista_vacia.append("Biología - Éxito")
        elif muestras >= 5 and temperatura <= 30:
            print("\nIA: Resultados aceptables.")
            print("Se recomienda continuar investigando.")
            lista_vacia.append("Biología - Aceptable")
        else:
            print("\nIA: Experimento fallido.")
            print("Las condiciones no son adecuadas.")
            lista_vacia.append("Biología - Fallido")

    # -------------------------
    # ÁREA TECNOLOGÍA
    # -------------------------
    elif area == "2":
        print("\nÁrea seleccionada: Tecnología")

        energia = int(input("Nivel de energía del robot (%): "))
        errores = int(input("Cantidad de errores detectados: "))

        # Uso de operadores relacionales
        if energia >= 80 and errores == 0:
            print("\nIA: El robot funciona perfectamente.")
            lista_vacia.append("Tecnología - Perfecto")
        elif energia >= 50 and errores <= 3:
            print("\nIA: El robot funciona con pequeñas fallas.")
            lista_vacia.append("Tecnología - Pequeñas Fallas")
        else:
            print("\nIA: Sistema en peligro.")
            print("Se necesita mantenimiento urgente.")
            lista_vacia.append("Tecnología - Peligro")

    # -------------------------
    # ÁREA ESPACIAL
    # -------------------------
    elif area == "3":
        print("\nÁrea seleccionada: Espacio")

        combustible = int(input("Cantidad de combustible: "))
        clima = input("¿El clima espacial es estable? (si/no): ")

        # Uso de operadores
        if combustible >= 100 and clima == "si":
            print("\nIA: Lanzamiento autorizado.")
            print("La misión espacial será un éxito.")
            lista_vacia.append("Espacio - Éxito")
        elif combustible >= 50 or clima == "si":
            print("\nIA: Lanzamiento con riesgo moderado.")
            lista_vacia.append("Espacio - Riesgo Moderado")
        else:
            print("\nIA: Misión cancelada.")
            print("Las condiciones son peligrosas.")
            lista_vacia.append("Espacio - Cancelada")

    # -------------------------
    # OPCIÓN INVÁLIDA
    # -------------------------
    else:
        print("\nIA: Opción no válida.")
        print("Reinicia el programa.")
        lista_vacia.append("Opción Inválida")

    print("\nHistorial de simulaciones: ", lista_vacia)
    continuar = input("\n¿Deseas realizar otra simulación? (si/no): ")
    if continuar.lower() != 'si':
        break

print("\nFin de la simulación.")
```
